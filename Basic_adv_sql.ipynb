```python
%%sql
#applying index inside partition for improving optimization
drop table employee_salary;
CREATE TABLE IF NOT EXISTS employee_salary (
    emp_id INT,
    employee_name VARCHAR(50),
    department VARCHAR(30),
    joining_month VARCHAR(20),
    salary INT,
    index idx_emp (employee_name),
    index idx_month(joining_month)
)

partition by list columns (department)(
    partition p_hr values in ('HR'),
    partition p_it values in ('IT'),
    partition p_fin values in ('Finance'),
    partition p_sal values in ('Sales')
)
```
