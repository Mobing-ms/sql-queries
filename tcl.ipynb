```python
%%sql
select * from purchase_details;
```

```python
%%sql
#similarly we can see year,month and day individually.
#add 30 days to current date
select date(order_date) + interval 30 day from purchase_details;
```

```python
%%sql
#formatting date
select
    p_id,
    product_name,
    date_format(order_date, '%W, %e %M %Y') as formatted_date 
from purchase_details;
```

```python
%%sql
#formatting time
select p_id,product_name,time_format(order_date,'%h:%i %p') from purchase_details;
```

```python
%%sql
#intoduction to regex expression
select * from purchase_details
where product_name regexp '^L';
```

```python
%%sql
select * from purchase_details
where product_name regexp '%n{2}%';
```
