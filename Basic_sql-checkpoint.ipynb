```python
%%sql
SELECT 
    p.club,
    (
        SELECT SUM(f.yellow_cards)
        FROM foul_attempted f
        WHERE f.player_id = p.player_id
    ) AS total_yellow_cards,
    (
        SELECT SUM(f.red_cards)
        FROM foul_attempted f
        WHERE f.player_id = p.player_id
    ) AS total_red_cards
    

FROM pl_footballers p;
```

```python
%%sql
create view club_fouls_attempted as
(
    SELECT 
    p.club,
    (
        SELECT SUM(f.yellow_cards)
        FROM foul_attempted f
        WHERE f.player_id = p.player_id
    ) AS total_yellow_cards,
    (
        SELECT SUM(f.red_cards)
        FROM foul_attempted f
        WHERE f.player_id = p.player_id
    ) AS total_red_cards
    FROM pl_footballers p
)
```

```python
%%sql
CREATE TABLE orders (
    order_id CHAR(3) PRIMARY KEY,
    product_name VARCHAR(40),
    amount INT,
    customer_id CHAR(3),
    FOREIGN KEY (customer_id) REFERENCES customers(customer_id)
);
```

```python

%%sql
#normal join or inner join
select 

c.customer_name,
o.product_name

from customers c
join orders o on c.customer_id = o.customer_id;
```

```python
%%sql
#left join 
select

c.customer_name,
coalesce(o.order_id,'No product') as order_id,
c.customer_id,
o.product_name

from customers c
left join orders o on c.customer_id = o.customer_id;
```

```python
%%sql
#right join
select

c.customer_name,
coalesce(c.customer_id,'nil') as customer_id,
o.order_id,
o.product_name

from orders o
right join customers c on c.customer_id = o.customer_id;
```
