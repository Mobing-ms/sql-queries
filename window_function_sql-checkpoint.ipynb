```python
%%sql
#this is the code to handle duplicate values using window function
with dupe_finder as(
    select *,row_number() over (partition by emp_id) as rows_number
from window_table
)

select emp_id,emp_name,department,salary,joining_year from dupe_finder where rows_number =1;
```

```python
%%sql

with dupe_finder as(
    select *,row_number() over (partition by emp_id) as rows_number
from window_table
)


select 

emp_id,
emp_name,
department,
salary,
dense_rank() over(order by salary desc) as emp_rank

from dupe_finder where rows_number = 1;
```

```python
%%sql

#the calculation of percent rank is nothing but rank of row -1/total number of rows -1

with dupe_finder as(
    select *,row_number() over (partition by emp_id) as rows_number
from window_table
)


select 

emp_id,
emp_name,
department,
salary,
percent_rank() over(order by salary desc) as emp_percent_rank

from dupe_finder where rows_number = 1;
```

```python
%%sql
#rank,dense_rank,ntile

with dupe_finder as(
    select *,row_number() over (partition by emp_id) as rows_number
from window_table
)


select 

emp_id,
emp_name,
department,
salary,
rank() over(order by salary desc) as emp_percent_rank,
ntile(4) over()

from dupe_finder where rows_number = 1;
```

```python
%%sql
#lag
select

employee_name,
department,
lag(sales_amount) over(partition by sale_id order by department) as previous_amount,
sales_amount - lag(sales_amount) over(partition by sale_id order by department) as differences

from monthly_sales;
```

```python
%%sql
#lead
select

employee_name,
department,
sales_amount,
lead(sales_amount) over(partition by sale_id order by department) as next_amount,
lead(sales_amount) over(partition by sale_id order by department) -sales_amount as differences

from monthly_sales;
```

```python
%%sql 
#last_value
select *,
last_value(sales_amount) over(partition by sale_id order by department rows between current row and unbounded following) as last__value

from monthly_sales;
```

```python
%%sql 
#first_value
select *,
first_value(sales_amount) over(partition by sale_id order by department) as first__value

from monthly_sales;
```

```python
%%sql

select *,
nth_value(sales_amount,1) over(partition by department order by sales_amount  desc ) as highest_sales
from monthly_sales;
```
